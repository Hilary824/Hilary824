```python
#2.3
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
demand=ctrl.Antecedent(np.arange(1,500.1,0.1),'demand')
competitor_p=ctrl.Antecedent(np.arange(5,15.1,0.1),'competitor_p')
production=ctrl.Antecedent(np.arange(50,500.1,0.1),'production')
seasonality=ctrl.Antecedent(np.arange(1,12.1,0.1),'seasonality')
stock=ctrl.Antecedent(np.arange(0,5000.1,0.1),'stock')
final_p=ctrl.Consequent(np.arange(5,20.1,0.1),'final_p')


demand['low']=fuzz.gaussmf(demand.universe,50,49)
demand['medium']=fuzz.gaussmf(demand.universe,200,150)
demand['high']=fuzz.gaussmf(demand.universe,500,300)

competitor_p['very low']=fuzz.gaussmf(competitor_p.universe,8,3)
competitor_p['low']=fuzz.gaussmf(competitor_p.universe,10,2)
competitor_p['medium']=fuzz.gaussmf(competitor_p.universe,15,5)
competitor_p['high']=fuzz.gaussmf(competitor_p.universe,20,5)

production['low']=fuzz.gaussmf(production.universe,150,100)
production['medium']=fuzz.gaussmf(production.universe,500,350)
production['high']=fuzz.gaussmf(production.universe,1000,500)

seasonality['off-peak']=fuzz.trimf(seasonality.universe,[1,6,6])
seasonality['normal']=fuzz.trimf(seasonality.universe,[6,6,8])
seasonality['peak']=fuzz.trimf(seasonality.universe,[6,8,12])

stock['low']=fuzz.gaussmf(stock.universe,1000,1000)
stock['medium']=fuzz.gaussmf(stock.universe,2000,1000)
stock['high']=fuzz.gaussmf(stock.universe,3500,1500)
stock['very high']=fuzz.gaussmf(stock.universe,5000,1500)

final_p['very low']=fuzz.gaussmf(final_p.universe,5.5,0.5)
final_p['low']=fuzz.gaussmf(final_p.universe,11.5,6)
final_p['medium']=fuzz.gaussmf(final_p.universe,13.5,2)
final_p['high']=fuzz.gaussmf(final_p.universe,15,1.5)
final_p['very high']=fuzz.gaussmf(final_p.universe,20,5)

rule1=ctrl.Rule(demand['high']&competitor_p['high']&stock['low'],final_p['very high'])
rule2=ctrl.Rule(demand['high']&seasonality['peak']&stock['low'],final_p['very high'])
rule3=ctrl.Rule(demand['medium']&production['high']&stock['low'],final_p['very high'])

rule4=ctrl.Rule(demand['high']&competitor_p['medium']&stock['medium'],final_p['high'])
rule5=ctrl.Rule(demand['medium']&competitor_p['high']&stock['low'],final_p['high'])
rule6=ctrl.Rule(demand['high']&seasonality['peak']&stock['medium'],final_p['high'])

rule7=ctrl.Rule(demand['medium']&competitor_p['medium']&stock['medium'],final_p['medium'])
rule8=ctrl.Rule(demand['low']&competitor_p['high']&stock['medium'],final_p['medium'])
rule9=ctrl.Rule(demand['medium']&production['medium']&stock['medium'],final_p['medium'])

rule10=ctrl.Rule(demand['low']&competitor_p['low']&stock['high'],final_p['low'])
rule11=ctrl.Rule(demand['medium']&production['low']&stock['high'],final_p['low'])
rule12=ctrl.Rule(demand['low']&seasonality['off-peak']&stock['high'],final_p['low'])

rule13=ctrl.Rule(demand['low']&competitor_p['low']&stock['very high'],final_p['very low'])
rule14=ctrl.Rule(demand['low']&seasonality['off-peak']&stock['very high'],final_p['very low'])
rule15=ctrl.Rule(demand['medium']&competitor_p['very low']&stock['very high'],final_p['very low'])

final_p.view()
```

```python
#2.5
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

temperature = ctrl.Antecedent(np.arange(16, 32, 1),'temperature')
humidity = ctrl.Antecedent(np.arange(40, 61, 1),'humidity')
people = ctrl.Antecedent(np.arange(1, 11, 1),'people')
fan = ctrl.Consequent(np.arange(740, 1251, 10),'fan')
cooling = ctrl.Consequent(np.arange(0, 101, 5),'cooling')

temperature['C'] = fuzz.trimf(temperature.universe, [16, 16, 19])
temperature['CL'] = fuzz.trimf(temperature.universe, [18, 20, 22])
temperature['CM'] = fuzz.trimf(temperature.universe, [21, 23, 25])
temperature['W'] = fuzz.trimf(temperature.universe, [24, 26, 28])
temperature['H'] = fuzz.trimf(temperature.universe, [27, 30, 32])

humidity['L'] = fuzz.trapmf(humidity.universe, [40, 40, 42, 45])
humidity['M'] = fuzz.trapmf(humidity.universe, [43, 45, 50, 52])
humidity['H'] = fuzz.trapmf(humidity.universe, [50, 55, 60, 60])

people['F'] = fuzz.gaussmf(people.universe, 2.5, 1)
people['M'] = fuzz.gaussmf(people.universe, 5, 1.5)
people['MA'] = fuzz.gaussmf(people.universe, 8, 1.5)

fan['L'] = fuzz.trapmf(fan.universe, [740, 740, 800, 880])
fan['M'] = fuzz.trapmf(fan.universe, [860, 900, 980, 1050])
fan['H'] = fuzz.trapmf(fan.universe, [1020, 1100, 1250, 1250])

cooling['L'] = fuzz.trimf(cooling.universe, [0, 10, 40])
cooling['M'] = fuzz.trimf(cooling.universe, [30, 50, 70])
cooling['H'] = fuzz.trimf(cooling.universe, [60, 80, 100])

rule1 = ctrl.Rule(temperature['C'], fan['L'])
rule2 = ctrl.Rule(temperature['CL'], fan['L'])
rule3 = ctrl.Rule(temperature['CM'], fan['M'])
rule4 = ctrl.Rule(temperature['W'], fan['H'])
rule5 = ctrl.Rule(temperature['H'], fan['H'])

rule6 = ctrl.Rule(temperature['C'], cooling['L'])
rule7 = ctrl.Rule(temperature['CL'], cooling['L'])
rule8 = ctrl.Rule(temperature['CM'], cooling['M'])
rule9 = ctrl.Rule(temperature['W'], cooling['H'])
rule10 = ctrl.Rule(temperature['H'], cooling['H'])

rule11 = ctrl.Rule(humidity['L'], cooling['L'])
rule12 = ctrl.Rule(humidity['M'], cooling['M'])
rule13 = ctrl.Rule(humidity['H'], cooling['H'])

rule14 = ctrl.Rule(people['F'], fan['L'])
rule15 = ctrl.Rule(people['M'], fan['M'])
rule16 = ctrl.Rule(people['MA'], fan['H'])

rule17 = ctrl.Rule(temperature['H'] & humidity['H'] & people['MA'], [cooling['H'], fan['H']])
rule18 = ctrl.Rule(temperature['CM'] & humidity['M'] & people['M'], [cooling['M'], fan['M']])
rule19 = ctrl.Rule(temperature['CL'] & humidity['L'] & people['F'], [cooling['L'], fan['L']])

AC_ctrl = ctrl.ControlSystem([
    rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11,
    rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19
    ])
AC_sim = ctrl.ControlSystemSimulation(AC_ctrl)
AC_sim.input['temperature'] = 25
AC_sim.input['humidity'] = 50
AC_sim.input['people'] = 5
AC_sim.compute()
print("Fan speed:", AC_sim.output['fan'])
print("Cooling level:", AC_sim.output['cooling'])
cooling.view(sim=AC_sim)
fan.view(sim=AC_sim)
```

```python
#2.6
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

weight = ctrl.Antecedent(np.arange(0, 1001, 1), 'weight')
food_type = ctrl.Antecedent(np.arange(0, 3.1, 0.1), 'food_type')
temp = ctrl.Antecedent(np.arange(0, 61, 1), 'temp')
time = ctrl.Consequent(np.arange(0, 21, 1), 'time')
power = ctrl.Consequent(np.arange(0, 101, 1), 'power')

weight['L'] = fuzz.trimf(weight.universe, [0, 200, 400])
weight['M'] = fuzz.trimf(weight.universe, [300, 500, 600])
weight['H'] = fuzz.trimf(weight.universe, [500, 900, 1000])

food_type['S'] = fuzz.trimf(food_type.universe, [0, 0.5, 1])
food_type['M'] = fuzz.trimf(food_type.universe, [0.8, 1.5, 2.2])
food_type['H'] = fuzz.trimf(food_type.universe, [1.8, 2.5, 3])

temp['C'] = fuzz.trimf(temp.universe, [0, 5, 10])
temp['W'] = fuzz.trimf(temp.universe, [20, 25, 30])
temp['H'] = fuzz.trimf(temp.universe, [40, 50, 60])

time['S'] = fuzz.trimf(time.universe, [0, 7, 12])
time['M'] = fuzz.trimf(time.universe, [5, 10, 12])
time['L'] = fuzz.trimf(time.universe, [10, 18, 20])

power['L'] = fuzz.trimf(power.universe, [0, 30, 40])
power['M'] = fuzz.trimf(power.universe, [30, 60, 70])
power['H'] = fuzz.trimf(power.universe, [60, 90, 100])

rule1 = ctrl.Rule(weight['L'], time['S'])
rule2 = ctrl.Rule(weight['M'], time['M'])
rule3 = ctrl.Rule(weight['H'], time['L'])

rule4 = ctrl.Rule(food_type['S'], power['L'])
rule5 = ctrl.Rule(food_type['M'], power['M'])
rule6 = ctrl.Rule(food_type['H'], power['H'])

rule7 = ctrl.Rule(temp['C'], time['L'])
rule8 = ctrl.Rule(temp['W'], time['M'])
rule9 = ctrl.Rule(temp['H'], time['S'])

rule10 = ctrl.Rule(weight['H'] & food_type['H'] & temp['C'], [time['L'], power['H']])
rule11 = ctrl.Rule(weight['M'] & food_type['S'] & temp['W'], [time['M'], power['L']])
rule12 = ctrl.Rule(weight['L'] & food_type['M'] & temp['H'], [time['S'], power['M']])

cooking_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
cooking_sim = ctrl.ControlSystemSimulation(cooking_ctrl)
cooking_sim.input['weight'] = 600
cooking_sim.input['food_type'] = 2
cooking_sim.input['temp'] = 10
cooking_sim.compute()
print("Thời gian nấu:",cooking_sim.output['time'])
print("Công suất:", cooking_sim.output['power'])
time.view(sim=cooking_sim)
power.view(sim=cooking_sim)
```

```python
#2.7
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl

food_type = ctrl.Antecedent(np.arange(0, 3.1, 0.1), 'food_type')
food_weight = ctrl.Antecedent(np.arange(0, 1001, 50), 'food_weight')
initial_temp = ctrl.Antecedent(np.arange(0, 61, 5), 'initial_temp')
water_content = ctrl.Antecedent(np.arange(0, 3.1, 0.1), 'water_content')
cook_time = ctrl.Consequent(np.arange(0, 31, 1), 'cook_time')
heat_level = ctrl.Consequent(np.arange(0, 101, 5), 'heat_level')

food_type['S'] = fuzz.trimf(food_type.universe, [0, 0.3, 1])
food_type['M'] = fuzz.trimf(food_type.universe, [0.8, 1.5, 2.2])
food_type['H'] = fuzz.trimf(food_type.universe, [1.8, 2.7, 3])

food_weight['L'] = fuzz.trimf(food_weight.universe, [0, 250, 500])
food_weight['M'] = fuzz.trimf(food_weight.universe, [400, 600, 800])
food_weight['H'] = fuzz.trimf(food_weight.universe, [700, 900, 1000])

initial_temp['C'] = fuzz.trimf(initial_temp.universe, [0, 5, 10])
initial_temp['W'] = fuzz.trimf(initial_temp.universe, [15, 25, 35])
initial_temp['H'] = fuzz.trimf(initial_temp.universe, [40, 50, 60])

water_content['L'] = fuzz.trimf(water_content.universe, [0, 0.4, 1])
water_content['M'] = fuzz.trimf(water_content.universe, [0.8, 1.5, 2.3])
water_content['H'] = fuzz.trimf(water_content.universe, [2, 2.6, 3])

cook_time['S'] = fuzz.trimf(cook_time.universe, [0, 7, 15])
cook_time['M'] = fuzz.trimf(cook_time.universe, [10, 20, 25])
cook_time['L'] = fuzz.trimf(cook_time.universe, [20, 25, 30])

heat_level['L'] = fuzz.trimf(heat_level.universe, [0, 20, 40])
heat_level['M'] = fuzz.trimf(heat_level.universe, [30, 50, 70])
heat_level['H'] = fuzz.trimf(heat_level.universe, [60, 80, 100])

rule1 = ctrl.Rule(food_type['S'] & food_weight['L'] & initial_temp['C'],
                  (cook_time['S'], heat_level['M']))
rule2 = ctrl.Rule(food_type['M'] & food_weight['M'] & initial_temp['W'],
                  (cook_time['M'], heat_level['M']))
rule3 = ctrl.Rule(food_type['H'] & food_weight['H'] & initial_temp['H'],
                  (cook_time['L'], heat_level['H']))
rule4 = ctrl.Rule(water_content['L'] & initial_temp['C'],
                  (cook_time['S'], heat_level['L']))
rule5 = ctrl.Rule(water_content['H'] & food_weight['H'],
                  (cook_time['L'], heat_level['H']))

cooking_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
cooking_sim = ctrl.ControlSystemSimulation(cooking_ctrl)

cooking_sim.input['food_type'] = 1
cooking_sim.input['food_weight'] = 500
cooking_sim.input['initial_temp'] = 25
cooking_sim.input['water_content'] = 2

cooking_sim.compute()

print("Cook time:", cooking_sim.output['cook_time'])
print("Heat level", cooking_sim.output['heat_level'])

cook_time.view(sim=cooking_sim)
heat_level.view(sim=cooking_sim)
```
