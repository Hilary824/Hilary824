```python
#2.3
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
demand=ctrl.Antecedent(np.arange(1,500.1,0.1),'demand')
competitor_p=ctrl.Antecedent(np.arange(5,15.1,0.1),'competitor_p')
production=ctrl.Antecedent(np.arange(50,500.1,0.1),'production')
seasonality=ctrl.Antecedent(np.arange(1,12.1,0.1),'seasonality')
stock=ctrl.Antecedent(np.arange(0,5000.1,0.1),'stock')
final_p=ctrl.Consequent(np.arange(5,20.1,0.1),'final_p')


demand['low']=fuzz.gaussmf(demand.universe,50,49)
demand['medium']=fuzz.gaussmf(demand.universe,200,150)
demand['high']=fuzz.gaussmf(demand.universe,500,300)

competitor_p['very low']=fuzz.gaussmf(competitor_p.universe,8,3)
competitor_p['low']=fuzz.gaussmf(competitor_p.universe,10,2)
competitor_p['medium']=fuzz.gaussmf(competitor_p.universe,15,5)
competitor_p['high']=fuzz.gaussmf(competitor_p.universe,20,5)

production['low']=fuzz.gaussmf(production.universe,150,100)
production['medium']=fuzz.gaussmf(production.universe,500,350)
production['high']=fuzz.gaussmf(production.universe,1000,500)

seasonality['off-peak']=fuzz.trimf(seasonality.universe,[1,6,6])
seasonality['normal']=fuzz.trimf(seasonality.universe,[6,6,8])
seasonality['peak']=fuzz.trimf(seasonality.universe,[6,8,12])

stock['low']=fuzz.gaussmf(stock.universe,1000,1000)
stock['medium']=fuzz.gaussmf(stock.universe,2000,1000)
stock['high']=fuzz.gaussmf(stock.universe,3500,1500)
stock['very high']=fuzz.gaussmf(stock.universe,5000,1500)

final_p['very low']=fuzz.gaussmf(final_p.universe,5.5,0.5)
final_p['low']=fuzz.gaussmf(final_p.universe,11.5,6)
final_p['medium']=fuzz.gaussmf(final_p.universe,13.5,2)
final_p['high']=fuzz.gaussmf(final_p.universe,15,1.5)
final_p['very high']=fuzz.gaussmf(final_p.universe,20,5)

rule1=ctrl.Rule(demand['high']&competitor_p['high']&stock['low'],final_p['very high'])
rule2=ctrl.Rule(demand['high']&seasonality['peak']&stock['low'],final_p['very high'])
rule3=ctrl.Rule(demand['medium']&production['high']&stock['low'],final_p['very high'])

rule4=ctrl.Rule(demand['high']&competitor_p['medium']&stock['medium'],final_p['high'])
rule5=ctrl.Rule(demand['medium']&competitor_p['high']&stock['low'],final_p['high'])
rule6=ctrl.Rule(demand['high']&seasonality['peak']&stock['medium'],final_p['high'])

rule7=ctrl.Rule(demand['medium']&competitor_p['medium']&stock['medium'],final_p['medium'])
rule8=ctrl.Rule(demand['low']&competitor_p['high']&stock['medium'],final_p['medium'])
rule9=ctrl.Rule(demand['medium']&production['medium']&stock['medium'],final_p['medium'])

rule10=ctrl.Rule(demand['low']&competitor_p['low']&stock['high'],final_p['low'])
rule11=ctrl.Rule(demand['medium']&production['low']&stock['high'],final_p['low'])
rule12=ctrl.Rule(demand['low']&seasonality['off-peak']&stock['high'],final_p['low'])

rule13=ctrl.Rule(demand['low']&competitor_p['low']&stock['very high'],final_p['very low'])
rule14=ctrl.Rule(demand['low']&seasonality['off-peak']&stock['very high'],final_p['very low'])
rule15=ctrl.Rule(demand['medium']&competitor_p['very low']&stock['very high'],final_p['very low'])

final_p.view()
```

```python
#2.4
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
load=ctrl.Antecedent(np.arange(1,500.1,0.1),'load')
dirt_level=ctrl.Antecedent(np.arange(8,15.1,0.1),'dirt_level')
fabric=ctrl.Antecedent(np.arange(50,500.1,0.1),'fabric')
time=ctrl.Consequent(np.arange(0,5000.1,0.1),'time')
water=ctrl.Consequent(np.arange(5,20.1,0.1),'water')


load['s']=fuzz.gaussmf(load.universe,50,49)
load['m']=fuzz.gaussmf(load.universe,200,150)
load['h']=fuzz.gaussmf(load.universe,500,300)

dirt_level['l']=fuzz.gaussmf(dirt_level.universe,10,2)
dirt_level['m']=fuzz.gaussmf(dirt_level.universe,15,5)
dirt_level['h']=fuzz.gaussmf(dirt_level.universe,20,5)

fabric['d']=fuzz.gaussmf(fabric.universe,150,100)
fabric['n']=fuzz.gaussmf(fabric.universe,350,200)
fabric['h']=fuzz.gaussmf(fabric.universe,500,150)

time['s']=fuzz.gaussmf(time.universe,1000,1000)
time['m']=fuzz.gaussmf(time.universe,2000,1000)
time['l']=fuzz.gaussmf(time.universe,5000,3000)

water['l']=fuzz.gaussmf(water.universe,11.5,6)
water['m']=fuzz.gaussmf(water.universe,13.5,2)
water['h']=fuzz.gaussmf(water.universe,15,1.5)

rule1=ctrl.Rule(load['s']&dirt_level['l'],time['s'],water['l'])
rule2=ctrl.Rule(load['s']&dirt_level['m'],time['m'],water['m'])
rule3=ctrl.Rule(load['s']&dirt_level['h'],time['l'],water['m'])

rule4=ctrl.Rule(load['m']&dirt_level['l'],time['m'],water['m'])
rule5=ctrl.Rule(load['m']&dirt_level['m'],time['l'],water['m'])
rule6=ctrl.Rule(load['m']&dirt_level['h'],time['l'],water['h'])

rule7=ctrl.Rule(load['h']&dirt_level['l'],time['m'],water['h'])
rule8=ctrl.Rule(load['h']&dirt_level['m'],time['l'],water['h'])
rule9=ctrl.Rule(load['h']&dirt_level['h'],time['l'],water['h'])

rule10=ctrl.Rule(fabric['d'],time['s'],water['l'])
rule11=ctrl.Rule(fabric['n'],time['m'],water['m'])
rule12=ctrl.Rule(fabric['h'],time['l'],water['h'])

rule13=ctrl.Rule(load['h']&dirt_level['h']&fabric['h'],time['l'],water['h'])
rule14=ctrl.Rule(load['s']&dirt_level['l']&fabric['d'],time['s'],water['l'])
rule15=ctrl.Rule(load['m']&dirt_level['m']&fabric['n'],time['m'],water['m'])

water.view()
```

```python
#2.5
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

temperature = ctrl.Antecedent(np.arange(5, 32, 1),'temperature')
humidity = ctrl.Antecedent(np.arange(30, 61, 1),'humidity')
people = ctrl.Antecedent(np.arange(1, 11, 1),'people')
fan = ctrl.Consequent(np.arange(800, 1251, 10),'fan')
cooling = ctrl.Consequent(np.arange(0, 101, 5),'cooling')

temperature['C'] = fuzz.trimf(temperature.universe, [5, 5, 14])
temperature['CL'] = fuzz.trimf(temperature.universe, [11, 18, 22])
temperature['CM'] = fuzz.trimf(temperature.universe, [21, 23, 24])
temperature['W'] = fuzz.trimf(temperature.universe, [25, 27, 28])
temperature['H'] = fuzz.trimf(temperature.universe, [30, 31, 32])

humidity['L'] = fuzz.trapmf(humidity.universe, [30, 30, 42, 45])
humidity['M'] = fuzz.trapmf(humidity.universe, [41, 47, 50, 52])
humidity['H'] = fuzz.trapmf(humidity.universe, [51, 57, 60, 60])

people['F'] = fuzz.gaussmf(people.universe, 2.5, 1)
people['M'] = fuzz.gaussmf(people.universe, 5, 1.5)
people['MA'] = fuzz.gaussmf(people.universe, 8, 1.5)

fan['L'] = fuzz.trapmf(fan.universe, [800, 800, 860, 880])
fan['M'] = fuzz.trapmf(fan.universe, [860, 900, 980, 1050])
fan['H'] = fuzz.trapmf(fan.universe, [1020, 1050, 1200, 1250])

cooling['L'] = fuzz.trimf(cooling.universe, [0, 30, 39])
cooling['M'] = fuzz.trimf(cooling.universe, [26, 58, 70])
cooling['H'] = fuzz.trimf(cooling.universe, [69, 90, 100])

rule1 = ctrl.Rule(temperature['C'], fan['L'])
rule2 = ctrl.Rule(temperature['CL'], fan['L'])
rule3 = ctrl.Rule(temperature['CM'], fan['M'])
rule4 = ctrl.Rule(temperature['W'], fan['H'])
rule5 = ctrl.Rule(temperature['H'], fan['H'])

rule6 = ctrl.Rule(temperature['C'], cooling['L'])
rule7 = ctrl.Rule(temperature['CL'], cooling['L'])
rule8 = ctrl.Rule(temperature['CM'], cooling['M'])
rule9 = ctrl.Rule(temperature['W'], cooling['H'])
rule10 = ctrl.Rule(temperature['H'], cooling['H'])

rule11 = ctrl.Rule(humidity['L'], cooling['L'])
rule12 = ctrl.Rule(humidity['M'], cooling['M'])
rule13 = ctrl.Rule(humidity['H'], cooling['H'])

rule14 = ctrl.Rule(people['F'], fan['L'])
rule15 = ctrl.Rule(people['M'], fan['M'])
rule16 = ctrl.Rule(people['MA'], fan['H'])

rule17 = ctrl.Rule(temperature['H'] & humidity['H'] & people['MA'], [cooling['H'], fan['H']])
rule18 = ctrl.Rule(temperature['CM'] & humidity['M'] & people['M'], [cooling['M'], fan['M']])
rule19 = ctrl.Rule(temperature['CL'] & humidity['L'] & people['F'], [cooling['L'], fan['L']])


cooling.view()
fan.view()
```
